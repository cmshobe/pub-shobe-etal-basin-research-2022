#### Script to produce Figure 3 in Shobe et al. (2022; _Basin Research_): Sediment flux supplied to basins as calculated by both basin-wide reconstructions and integration of individual seismic sections.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Import sediment fluxes calculated by integrating individual
seismic sections. These calculations are done in the 
preprocessing ("prepro") script for each section.

```python
qs_r1 = np.load('../marine/sections/orange_section_R1/prepro/qs_array.npy')
qs_r3 = np.load('../marine/sections/orange_section_R3/prepro/qs_array.npy')
qs_r4 = np.load('../marine/sections/orange_section_R4/prepro/qs_array.npy')
qs_r5 = np.load('../marine/sections/orange_section_R5/prepro/qs_array.npy')
qs_r6 = np.load('../marine/sections/orange_section_R6/prepro/qs_array.npy')
qs_r7 = np.load('../marine/sections/orange_section_R7/prepro/qs_array.npy')
qs_r8 = np.load('../marine/sections/orange_section_R8/prepro/qs_array.npy')
```

Import data from basin-wise reconstructions as reported in Baby et al. (2019; _Terra Nova_)

```python
#walvis basin
walvis_qs = np.array([4.1,
1.3,
0.6,
2.2,
7.1,
7.5,
4.6,
0.6])[::-1]

walvis_err = np.array([1.0,
0.3,
0.1,
0.5,
1.6,
1.9,
1.2,
0.2])[::-1]

#luderitz basin

luderitz_qs = np.array([3.3,
0.5,
0.3,
2.4,
6.8,
10.7,
0.9,
0.07])[::-1]

luderitz_err = np.array([0.8,
0.1,
0.06,
0.5,
1.5,
2.6,
0.2,
0.02])[::-1]

#orange basin

orange_qs = np.array([6.2,
2.4,
0.6,
6.1,
51.4,
19.9,
4.6,
2.7])[::-1]

orange_err = np.array([1.4,
0.6,
0.1,
1.3,
11.0,
4.9,
1.1,
1.0])[::-1]

#cape basin

cape_qs = np.array([19.2,
1.8,
0.04,
14.4,
0.3,
4.3,
0.8,
0.4])[::-1]
cape_err = np.array([4.3,
0.5,
0.01,
3.1,
0.06,
1.1,
0.2,
0.1])[::-1]
```

Convert data from Baby et al. (2019) into units of m$^3$/yr

```python
layer_starts = [0, 17000000, 30000000, 36000000, 49000000, 64000000, 100000000, 119000000]
layer_durations = [17000000, 13000000, 6000000, 13000000, 15000000, 36000000, 19000000, 11000000]

#walvis basin
walvis_qs_array = np.zeros(130000)
for i in range(8): #for each layer
    layer_start = int(layer_starts[i] / 1000)
    layer_duration = int(layer_durations[i] / 1000)
    
    qs = walvis_qs[i]

    walvis_qs_array[layer_start:layer_start + layer_duration] = qs * 1000 / 1000000 * 1e9
    #conversion factor is *1000 to go from 10^3 km^3/Myr to km^3/Myr, then
    #/1000000 to go from km^3/Ma to km^3/yr, then
    #*1e9 to go from km^3/yr to m^3/yr
    
#luderitz basin
luderitz_qs_array = np.zeros(130000)
for i in range(8): #for each layer
    layer_start = int(layer_starts[i] / 1000)
    layer_duration = int(layer_durations[i] / 1000)
    
    qs = luderitz_qs[i]

    luderitz_qs_array[layer_start:layer_start + layer_duration] = qs * 1000 / 1000000 * 1e9
    
#orange basin
orange_qs_array = np.zeros(130000)
for i in range(8): #for each layer
    layer_start = int(layer_starts[i] / 1000)
    layer_duration = int(layer_durations[i] / 1000)
    
    qs = orange_qs[i]

    orange_qs_array[layer_start:layer_start + layer_duration] = qs * 1000 / 1000000 * 1e9
    
#cape basin
cape_qs_array = np.zeros(130000)
for i in range(8): #for each layer
    layer_start = int(layer_starts[i] / 1000)
    layer_duration = int(layer_durations[i] / 1000)
    
    qs = cape_qs[i]

    cape_qs_array[layer_start:layer_start + layer_duration] = qs * 1000 / 1000000 * 1e9
```

Make Figure 3

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 22})
matplotlib.rcParams.update({'legend.title_fontsize': 14})

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(6, 10))

#UPPER PLOT: TIME SERIES FROM BABY ET AL 2019
ax1.plot(walvis_qs_array/1e6, linewidth = 3, label = 'Walvis', color ='#e41a1c')
ax1.plot(luderitz_qs_array/1e6, linewidth = 3, label = 'Luderitz', color ='#377eb8')
ax1.plot(orange_qs_array/1e6, linewidth = 3, label = 'Orange', color ='#4daf4a')
ax1.plot(cape_qs_array/1e6, linewidth = 3, label = 'Cape', color ='#984ea3')

ax1.legend(title = 'Basin', fontsize = 14)

ax1.set_ylabel('Volumetric sediment' '\n' r'flux [$\times 10^6 $m$^3$/yr]')
ax1.set_ylim(-1, 60)
ax1.text(1000, 55, 'A) Basin-integrated')

#LOWER PLOT: TIME SERIES EXTRACTED FROM INTEGRATION OF INDIVIDUAL SECTIONS

ax2.plot(qs_r1, linewidth = 3, label = '1', color = '#1b9e77')
ax2.plot(qs_r3, linewidth = 3, label = '3', color = '#d95f02')
ax2.plot(qs_r4, linewidth = 3, label = '4', color = '#7570b3')
ax2.plot(qs_r5, linewidth = 3, label = '5', color = '#e7298a')
ax2.plot(qs_r6, linewidth = 3, label = '6', color = '#66a61e')
ax2.plot(qs_r7, linewidth = 3, label = '7', color = '#e6ab02')
ax2.plot(qs_r8, linewidth = 3, label = '8', color = '#a6761d')


ax2.set_xlim(0, 130000)
ax2.set_ylim(-1, 100)
ax2.legend(title = 'Section', fontsize = 14)

ax2.xaxis.set_ticks(np.arange(0, 130000, 20000))

a=ax2.get_xticks().tolist()
a[:] = [130, 110, 90, 70, 50, 30, 10]
ax2.set_xticklabels(a)

ax2.set_xlabel('Time before present [Ma]')
ax2.set_ylabel('Sediment flux per''\n' 'unit basin width [m$^2$/yr]')

ax2.text(1000, 91, 'B) Individual sections')
```

```python
fig.savefig('Shobe_etal_fig3.png', dpi=1000, bbox_inches='tight', facecolor='white', transparent=False)
```
